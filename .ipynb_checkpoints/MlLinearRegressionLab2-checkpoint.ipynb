```python
import numpy as np
from numpy.linalg import pinv
from matplotlib import pyplot
```

```python
def read_array_from_file(file):
    f = list(open(file, 'r'))
    n = int(f[0])
    m = int(f[1])
    train_data = np.array([[int(j) for j in np.array(f[i + 2].split(' '))] for i in range(m)])
    k = int(f[m + 2])
    test_data = np.array([[int(j) for j in np.array(f[i + m + 3].split(' '))] for i in range(k)])
    return n, train_data, test_data
```

```python
n, train_data, test_data = read_array_from_file('LR/1.txt')
```

```python
def predict_pseudoinverse(train_data, test_data, n):
    X_train, y_train = train_data[:,:-1], train_data[:,n]
    X_train = X_train.reshape((len(X_train), n))
    b = pinv(X_train).dot(y_train)
    X_test, y_test = test_data[:,:-1], test_data[:,n]
    y_predicted = X_test.dot(b)
#     print(y_test)
#     print(y_predicted)
    pyplot.plot(y_predicted, label='prediction')
    pyplot.plot(y_test, label='targey')
    pyplot.legend()
    pyplot.show()

```

```python
predict_pseudoinverse(train_data, test_data, n)
```

```python
def predict_gradient(train_data, test_data, n, iterations):
    costs = []
    X_train, y_train = train_data[:,:-1], train_data[:,n]
    theta = np.ones(n)
    alpha = 0.000000000000001
    for i in range(0, iterations):
        hypothesis = np.dot(X_train, theta)
        loss = hypothesis - y_train
        # avg cost per example (the 2 in 2*m doesn't really matter here.
        # But to be consistent with the gradient, I include it)
        cost = np.sum(loss ** 2) / (2 * n)
        print("Iteration %d | Cost: %f" % (i, cost))
        # avg gradient per example
        gradient = np.dot(X_train.T, loss) / n
        # update
        theta = theta - alpha * gradient
    return theta
#     X_train = X_train.reshape((len(X_train), n))
#     w = np.random.randn(n) / np.sqrt(n)
#     learning_rate = 0.001
#     for t in range(1000):
#         y_predicted = X_train.dot(w)
#         delta = y_predicted - y_train
#         w = w - learning_rate * X_train.T.dot(delta)
#         mse = delta.dot(delta) / n
#         costs.append(mse)
#     pyplot.plot(costs)
#     pyplot.show()

```

```python
print(predict_gradient(train_data, test_data, n, 1000))
```

```python
def draw_rmse(train_data, test_data, n):
    res = np.array([])
    for i in range(100, 500, 100):
        theta = predict_gradient(train_data, test_data, n, i)
        X_test, y_test = test_data[:,:-1], test_data[:,n]
        y_predicted = X_test.dot(theta)
        y = y_test - y_predicted
#         y = [i ** 2 for i in y]
        np.concatenate(res, [sum(y)])
    print(res)
    pyplot.plot(range(100, 500, 100), res)
    pyplot.show()
        
```

```python
draw_rmse(train_data, test_data, n)
```
